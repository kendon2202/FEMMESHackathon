# Editing Image Pixels

We will use the Image module within the larger Python Imaging Library, called PIL. The documentation for the Image module is here: http://effbot.org/imagingbook/image.htm

The following example is adapted from https://en.wikibooks.org/wiki/Python_Imaging_Library/Editing_Pixels

```
import matplotlib.pyplot as plt

from PIL import Image # imports the Image library from PIL. 
# We don't need to take up extra memory by importing all PIL libraries
# so we just import Image.

```

```
# PIL accesses images in Cartesian co-ordinates, so it is Image[columns, rows]
import matplotlib.pyplot as plt

from PIL import Image
image_size = (250,250)
img = Image.new('RGB', image_size, "black") # create a new black image
pixelMap = img.load() # create the pixel map of img, currently a black square.

plt.imshow(img)
```

```
for i in range(img.size[0]): # for every column
  for j in range(img.size[1]): # for every row
    pixelMap[i,j] = (i,j,50)  # set the color of the [i,j] coordinate pixel
plt.imshow(img)
print('\n')
for i in range(img.size[0]): # for every column
  for j in range(img.size[1]): # for every row
    pixelMap[i,j] = (i,j,10)  # set the color of the [i,j] coordinate pixel

plt.imshow(img)
print('\n')
for i in range(img.size[0]): # for every column
  for j in range(img.size[1]): # for every row
    pixelMap[i,j] = (i,j,150)  # set the color of the [i,j] coordinate pixel

plt.imshow(img)
```

The line pixelMap[i,j] = (i,j,100) shows the structure of an image. Each pixel is located at the index [i,j] in the 2-D array, pixelMap. This pixel has an associated **tuple** containing three items (similar to a list but parenthesis instead of square brackets).

The tuple corresponds to intensity of colors in that pixel. The three colors are Red, Green, and Blue. By convention, the minimum intensity value of a color is 0, and the maximum intensity value of a color is 255.
